```python
#Importing required libraries
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
#Step-2: Reading the image

img = cv2.imread("/content/albert.jpg")
from skimage import io 
io.imshow(img)
```

```python
##Converting to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
io.imshow(img)
```

```python
#Step-3: Detecting edges
#Detecting edges of the input image
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.medianBlur(gray, 5)
edges = cv2.adaptiveThreshold(gray, 150, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)
io.imshow(edges)
```

```python
#Step-4: Cartoonifying image
#cartoonifying the image using bilateral filter method
#Cartoonifying the image
color = cv2.bilateralFilter(img, 9, 250, 250)
cartoon = cv2.bitwise_and(color, color, mask=edges)
```

```python
#Step-5: Final Output (Cartoon Image)
io.imshow(cartoon)
```
